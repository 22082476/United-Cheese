# Frequent itemsets with Apyori

```python
from apyori import apriori
import pandas as pd
import warnings
import pyodbc
warnings.simplefilter("ignore")
```

## Boodschappen

### Data kneden

```python
data = [["Melk", "Brood", "Eieren"], ["Appels", "Bananen", "Sinaasappels"], ["Rijst", "Pasta", "Tomaten"], ["Kipfilet", "Zalm", "Biefstuk"], ["Yoghurt", "Kaas", "Boter"], ["Spinazie", "Wortelen", "Komkommer"], ["Melk", "Appels", "Kipfilet"], ["Brood", "Kaas", "Yoghurt"], ["Eieren", "Tomaten", "Melk"], ["Bananen", "Kaas", "Pasta"]]
```

```python
rules = apriori(data, min_support=0.1, min_confidence=0.6, min_lift=1.0, min_length=2)


for rule in rules:
    items = [item for item in rule.items]
    antecedent = ', '.join(items[:-1])
    consequent = items[-1]
    confidence = rule.ordered_statistics[0].confidence
    support = rule.support
    print(f"Regel: {antecedent}(X) => {consequent}(Y), Vertrouwen(Bij X hoevaak Y): {confidence:.2f}, Ondersteuning(Hoevaak komt X voor): {support:.2f}")
```

```python


DB = {'servername' : 'MSI\\SQLEXPRESS',
      'database' : 'GreatOutdoorsDatawareHouse1'}
export_conn = pyodbc.connect('DRIVER={SQL server};SERVER=' + DB['servername'] + ';DATABASE=' + DB['database'] + ';Trusted_Connection=yes')
   

data = pd.read_sql_query("SELECT * FROM order_table", export_conn)
product = pd.read_sql_query("SELECT * FROM product_table", export_conn)
data
```

```python
product = product.loc[:, ["product_number", "product_name"]]
product["product_number"] = product["product_number"].astype(str)
data = data.loc[:, ["order_number", "product_number"]]
data
```

```python
def makeToApyoriList (data):
    orders = data.loc[:, ["order_number"]].drop_duplicates()
    resultList = []
    print(len(orders))
    for index, row in orders.iterrows():
        orderList = []
        for index, orderItem in data.loc[(data["order_number"] == row["order_number"]), ["product_number"]].iterrows():
            orderList.append(orderItem["product_number"].astype(str))
        resultList.append(orderList)

    return resultList

data = makeToApyoriList(data)
data
```

```python
rules = apriori(data, min_support=0.02, min_confidence=0.3, min_lift=1.0, min_length=1)

def getProductName (items):
    productList = []

    for item in items:
        name = product.loc[(product["product_number"] == item), "product_name"].iloc[0]
        productList.append(name)

    return productList

for rule in rules:
    items = getProductName(rule.items)
    antecedent = ', '.join(items[:-1]) if len(items) > 1 else items[0]
    consequent = items[-1]  # Consequent is het laatste item
    confidence = rule.ordered_statistics[0].confidence
    support = rule.support
    print(f"Regel: {antecedent}(X) => {consequent}(Y), Vertrouwen(Bij X hoevaak Y): {confidence:.2f}, Ondersteuning(Hoevaak komt X voor): {support:.2f}")
```
